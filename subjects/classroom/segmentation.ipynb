<h3>Q1: Segmentar o texto</h3>

```python
import cv2
import matplotlib.pyplot as plt

def segmentation(img):
    image = img

    image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) 
    
    _, global_threshold = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)

    _, otsu_threshold = cv2.threshold(image, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

    titles = ['Original', 'Global Threshold', 'Otsu Threshold']
    images = [image, global_threshold, otsu_threshold]

    plt.figure(figsize=(15, 5))

    for i in range(3):
        plt.subplot(1, 3, i+1)
        plt.imshow(images[i], 'gray')
        plt.title(titles[i])
        plt.xticks([]), plt.yticks([])

    plt.show()

# Example usage:
# Assuming img is your input image
# segmentation(img)
```

```python
segmentation(img=img)
```

```python
img  = cv2.imread('../../images/building.tif', 1)
plt.imshow(img, cmap='gray')  
```

```python
def find_lines(img):

    gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    edges = cv2.Canny(gray_image,50,150,apertureSize=3)
    
   
    lines_list =[]
    lines = cv2.HoughLinesP(
                edges,  
                1,  
                np.pi/180,  
                threshold=100,  
                minLineLength=5,  
                maxLineGap=10 
                )
    
    # Iterate over points
    for points in lines:
        x1,y1,x2,y2=points[0]
        cv2.line(img,(x1,y1),(x2,y2),(0,255,0),2)
        lines_list.append([(x1,y1),(x2,y2)])
        
    plt.imshow(img)
```

```python
img_copy = img

find_lines(img=img_copy)
```
