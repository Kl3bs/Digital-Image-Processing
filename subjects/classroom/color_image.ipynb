```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
IMG_PATH = '../../images'
img = cv2.imread(f"{IMG_PATH}/chips.png" )
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img)
```

<h3>YcrCb Convertion</h3>

```python
ycr_img = cv2.cvtColor(img, cv2.COLOR_BGR2YCrCb)
plt.imshow(ycr_img)
```

```python
def plot_channels(c1,c2,c3):
    #Plotting the color channels
    plt.figure(figsize=(12,12))
    plt.subplot(131),plt.imshow(c1)
    plt.title('C1')
    plt.subplot(132),plt.imshow(c2)
    plt.title('C2')
    plt.subplot(133),plt.imshow(c3)
    plt.title('C3')
    plt.show()


c1,c2,c3 =  cv2.split(ycr_img)
plot_channels(c1,c2,c3)
```

```python
#Plotting histogram by channel

def plot_ycrcb_hist(img):
    # Splitting the RGB channels into variables
    y,cr,cb = cv2.split(img)
    
    # Plotting each channel histogram
    plt.figure(figsize=(12,4))

    plt.subplot(141)
    plt.title('Y')
    plt.hist(y.ravel(),bins=256,range=[0,256], color='y')

    plt.subplot(142)    
    plt.title('Cr')
    plt.hist(cr.ravel(),bins=256,range=[0,256], color='r')

    plt.subplot(143)
    plt.title('Cb')
    plt.hist(cb.ravel(),bins=256,range=[0,256], color='b')

    plt.xlim([0, 256])
    plt.show()

plot_ycrcb_hist(ycr_img)
```

```python
y,cr,cb = cv2.split(ycr_img)

# Set Y to 0
y1 = np.zeros(y.shape, dtype=y.dtype)
cr1 = cr
cb1 = cb
img1 = cv2.merge((y1, cr1, cb1))

# Set Cr to 0
y2 = y  
cr2 = np.zeros(cr.shape, dtype=cr.dtype)
cb2 = cb
img2 = cv2.merge((y2, cr2, cb2)) 

# Set Cb to 0
y3 = y
cr3 = cr
cb3 = np.zeros(cb.shape, dtype=cb.dtype)  
img3 = cv2.merge((y3, cr3, cb3))

# Plot images
plt.figure(figsize=(15,15))
plt.subplot(1,3,1)
plt.imshow(img1)
plt.title('Y = 0')

plt.subplot(1,3,2) 
plt.imshow(img2)
plt.title('Cr = 0')

plt.subplot(1,3,3)
plt.imshow(img3)
plt.title('Cb = 0')

plt.show()
```

<h3>HSV</h3>

```python
hsv_img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
plt.imshow(hsv_img)
```

```python
h,s,v = cv2.split(hsv_img)
plot_channels(h,s,v)
```

```python
def plot_hsv_hist(img):
    # Splitting the RGB channels into variables
    c1,c2,c3 = cv2.split(img)
    
    # Plotting each channel histogram
    plt.figure(figsize=(12,4))

    plt.subplot(141)
    plt.title('Hue')
    plt.hist(c1.ravel(),bins=256,range=[0,256], color='brown')

    plt.subplot(142)    
    plt.title('Saturation')
    plt.hist(c2.ravel(),bins=256,range=[0,256], color='b')

    plt.subplot(143)
    plt.title('Value')
    plt.hist(c3.ravel(),bins=256,range=[0,256], color='pink')

    plt.x
    plt.xlim([0, 256])
    plt.show()

plot_hsv_hist(hsv_img)
```
