```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
#Task: remove the image green image background
IMG_PATH = "../../images/avengers.png"

img = cv2.imread(IMG_PATH)
# img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img)
```

```python
def remove_green_background(image):
    # Convert BGR to HSV colorspace 
    hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

    # Define range of green color in HSV
    lower_green = np.array([40, 40, 40]) 
    
    upper_green = np.array([140, 255, 255])

    # Create mask to only select green
    mask = cv2.inRange(hsv, lower_green, upper_green)

    # Change pixels in mask to black, others to white
    mask = cv2.bitwise_not(mask)

    # Convert BGR to RGBA colorspace
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGBA)

    # Use mask to create transparent layer
    image_rgb[:, :, 3] = mask 

    return image_rgb
```

```python
res = remove_green_background(img)
plt.imshow(res)
```
