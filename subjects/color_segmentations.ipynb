```python
#Plotting each channel histogram

def plot_hist(img):
    # Splitting the RGB channels into variables
    r,g,b = cv2.split(img)
    
    # Plotting each channel histogram
    plt.figure(figsize=(7,7))
    plt.hist(r.ravel(),bins=256,range=[0,256], color='r')
    plt.hist(g.ravel(),bins=256,range=[0,256], color='g')
    plt.hist(b.ravel(),bins=256,range=[0,256], color='b')
    plt.show()

plot_hist(rgb_img2)
```

```python
#Getting HSV image 
hsv_img = cv2.cvtColor(rgb_img2, cv2.COLOR_BGR2HSV)

plt.imshow(hsv_img)
```

```python
#Plotting Hue, Saturation and Value from image

def hsv_plot(img):
    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    h, s, v = cv2.split(hsv)
    plt.figure(figsize=(7,7))
    plt.scatter(h.ravel(),s.ravel(),c=v.ravel())
    plt.show()

hsv_plot(rgb_img2)
```

```python
def hsv_hist_plot(image):
    # Convert BGR to HSV
    hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

    # Plotting the histogram
    hist = cv2.calcHist([hsv], [0, 1], None, [180, 256], [
        0, 180, 0, 256])  # [0,180,0,256]
    
    # Plotting the histogram
    plt.figure(figsize=(10, 10))
    plt.plot(hist)

hsv_hist_plot(rgb_img2)
```

```python
#Plot image in grayscale

def plot_in_grayscale(img):
    gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    plt.imshow(gray_image, cmap='gray')
    plt.title('Grayscale Image')
    plt.axis('off')  
    plt.show()

plot_in_grayscale(hsv_img)
```
